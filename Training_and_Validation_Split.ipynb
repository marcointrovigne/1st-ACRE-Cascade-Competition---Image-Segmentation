```
# Split Training and Validation
firm = 'Roseau'
plant = 'Haricot'
RESET_VALIDATION = False                   
DATASET_SPLIT = 0.85
```

```
path_training = '/content/drive/MyDrive/Kaggle_Competition/Kaggle_2/Development_Dataset_Split/training'
path_validation = '/content/drive/MyDrive/Kaggle_Competition/Kaggle_2/Development_Dataset_Split/validation'

if RESET_VALIDATION:
    if not os.path.exists(path_training + '/' + firm + '/' + plant) and not os.path.exists(path_validation + '/' + firm + '/' + plant):
        if not os.path.exists(path_training + '/' + firm) and not os.path.exists(path_validation + '/' + firm):
            os.mkdir(path_training + '/' + firm)
            os.mkdir(path_validation + '/' + firm)
        os.mkdir(path_training + '/' + firm + '/' + plant)
        os.mkdir(path_validation + '/' + firm + '/' + plant)                                                                         
    if os.path.exists(path_training + '/' + firm + '/' + plant) and os.path.exists(path_validation + '/' + firm + '/' + plant):
        shutil.rmtree(path_training + '/' + firm + '/' + plant)
        os.mkdir(path_training + '/' + firm + '/' + plant)
        shutil.rmtree(path_validation + '/' + firm + '/' + plant)
        os.mkdir(path_validation + '/' + firm + '/' + plant) 

    # Destination paths
    dest_train = path_training + '/' + firm + '/' + plant
    dest_valid = path_validation + '/' + firm + '/' + plant
    
    if not os.path.exists(dest_train+'/'+'Images'+'/img'):
        os.mkdir(dest_train+'/'+'Images')
        os.mkdir(dest_train+'/'+'Images'+'/img')
    if not os.path.exists(dest_valid+'/'+'Images'+'/img'):
        os.mkdir(dest_valid+'/'+'Images')
        os.mkdir(dest_valid+'/'+'Images'+'/img')    
    if not os.path.exists(dest_train+'/'+'Masks'+'/img'):
        os.mkdir(dest_train+'/'+'Masks')
        os.mkdir(dest_train+'/'+'Masks'+'/img')
    if not os.path.exists(dest_valid+'/'+'Masks'+'/img'):
        os.mkdir(dest_valid+'/'+'Masks')
        os.mkdir(dest_valid+'/'+'Masks'+'/img')

    source_images = '/content/drive/MyDrive/Kaggle_Competition/Kaggle_2/Development_Dataset/Training' + '/' + firm + '/' + plant +'/Images'
    source_masks = '/content/drive/MyDrive/Kaggle_Competition/Kaggle_2/Development_Dataset/Training' + '/' + firm + '/' + plant +'/Masks'     

    # Dataset split
    dataset_split = {}  
    dataset_split["training"] = []
    dataset_split["validation"] = []

    files_img = os.listdir(source_images)
    random.seed(SEED)
    random.shuffle(files_img)

    # Create training set randomly
    for i in range(int(len(files_img)*DATASET_SPLIT)):
        dest = shutil.copy(source_images + '/'+ files_img[i], dest_train + '/Images/img/' + files_img[i])
        dest = shutil.copy(source_masks + '/'+ files_img[i][:-3] + 'png', dest_train + '/Masks/img/' + files_img[i][:-3] + 'png')
        dataset_split["training"].append(files_img[i])

    # Create validation set randomly
    for j in range(i+1, len(files_img)):
        dest = shutil.copy(source_images + '/' + files_img[j], dest_valid + '/Images/img/' + files_img[j])
        dest = shutil.copy(source_masks + '/' + files_img[j][:-3] + 'png', dest_valid + '/Masks/img/' + files_img[j][:-3] + 'png')
        dataset_split["validation"].append(files_img[j])
else:
    dest_train = path_training + '/' + firm + '/' + plant
    dest_valid = path_validation + '/' + firm + '/' + plant

print("Images uploaded for training:", len(os.listdir(dest_train+'/Images/img')))
print("Images uploaded for validation:", len(os.listdir(dest_valid+'/Images/img')))
```
